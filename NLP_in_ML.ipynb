```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


df = pd.read_csv('train.txt',sep = ';',header = None,names = ['text','emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#  NOW HERE WE ARE FINDING UNIQUE EMOTIONS
unique_emotions = df['emotion'].unique()
#  HERE WE HAVE ARRAY IN WHICH WE STORE EMOTIONS
emotion_numbers = {}
#  STARTING FROM I =0
i =0
# HERE WE ARE APPLYING LOOP AND ADDING UNIQUE EMOTIONS IN ARRAY
for emo in unique_emotions:
  emotion_numbers[emo]=i
  # MAKING COUNT INCREASE FOR TAKING MORE UNIQUE EMOTIONS
  i+=1

#  AND HERE WE ARE MAPPING EMOTION ARRAY
df['emotion'] = df['emotion'].map(emotion_numbers)
```

```python
df
```

```python
df['text']= df['text'].apply(lambda x : x.lower())
```

```python
#  HERE WE ARE REMOVING PUNCTUAUTIONS
#  HERE WE BASICALLY MAKE A FUNCTION
import string

def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
#  HERE WE ARE APPLYING THE PUNC FUNCTION
df['text'] = df['text'].apply(remove_punc)
```

```python
#  NOW HERE WE ARE REMOVING NUMBERS FROM THE TEXT
def remove_numbers(txt):
  new = ""
  for i in txt:
    if not i.isdigit():
      new+=i
  return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
#  HERE WE HAVE FUNC FOR REMOVING THE EMOJIS FROM THE TEXT
def remove_emojis(txt):
  new = ""
  for i in txt:
    if i.isascii():
      new+=i
  return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
#  NOW WE ARE IMPORTING THE SPECIFIC NPL LIBRARY
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python

df.loc[1]['text']
```

```python
#  HERE WE HAVE CODE FOR REMOVING STOP_WORDS FROM THE TEXT
def remove(txt):
  words = txt.split()
  cleaned = []
  for i in words:
    # BY DEFAULT ENGLISH KE STOPWARDS PRE-DEFINED REHTE HAI
    # AGAR WO PRE-DEFINED WORDS STOPWARDS ME JO REHTE HAI WO NHI HAI TO CLEANED ARRAY ME DALDEGE(APPEND)
    if not i in stop_words:
      cleaned.append(i)
  return ' '.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
```

```python
#  NOW WE ARE IMPORTING SKLEARN FOR VARIOUS MODEL TEST

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.20, random_state=42)
```

```python
#  THIS IS BASICALLLY BAG-OF-WORDS AND TF-IDF IMPORTING
#  THIS ARE BASICALLY OLD AND COMPLEX TECHNIQUES TO PREDICT ACCURACY IN NLP
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

#  ALSO WE ARE APPLYING NAIVE BAYES
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)


nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)


pred_bow = nb_model.predict(X_test_bow)
print(accuracy_score(y_test, pred_bow))
```

```python
pred_bow
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf, y_train)

y_pred = nb2_model.predict(X_test_tfidf)
print(accuracy_score(y_test, y_pred))
```

```python
#  NOW WE ARE USING LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
```
